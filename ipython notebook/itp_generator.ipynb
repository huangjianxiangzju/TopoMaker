```python
angle=[]
for i in range(262):
    a=conect[i][0]
    b=conect[i][1]
    for j in range(len(conect)):
        c=conect[j][0]
        d=conect[j][1]
        if ( a ==c ) and (b != c and b!=d):
            if [d,a,b] not in angle:
                angle.append([b,a,d])
        elif (a==d) and (b != c and b!=d):
            if [c,a,b] not in angle:
                angle.append([b,a,c])
        elif ( b ==c ) and (a != c and a!=d):
            if [d,b,a] not in angle:
                angle.append([a,b,d])
        elif (b==d) and (a != c and a!=d):
            if [c,b,a] not in angle:
                angle.append([a,b,c])
```
