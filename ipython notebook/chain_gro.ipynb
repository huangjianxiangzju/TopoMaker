```python
import MDAnalysis as mda
```

```python
struct=mda.Universe("pei_mono.pdb")
```

```python
H1=struct.select_atoms("name H1")
H2=struct.select_atoms("name H2")
H3=struct.select_atoms("name H3")
H4=struct.select_atoms("name H4")
```

```python
import matplotlib.pyplot as plt
```

```python
all=struct.select_atoms("all")
```

```python
pos=all.positions
```

```python
plt.scatter(pos[:,0],pos[:,1],s=200)
```

```python
vec1=(H1.positions-H2.positions)
vec2=(H3.positions-H4.positions)
```

```python
listname=['H1','H2','H3','H4']
```

```python
import numpy as np
check=[]
pos_new=np.zeros_like(pos)
previous=''
pos_cum=np.zeros((26*15,3))
pos_cum[0:15,:]=pos
i=0
while i<25:
    name=listname[np.random.randint(0,4)]
    print(name)

    if name=='H1' and previous!='H2':
        pos_new=pos+vec1
        #pos_cum=np.vstack((pos_new,pos))
        pos=pos_new
        previous='H1'
        plt.scatter(pos_new[:,0],pos_new[:,1],s=20,c='r')
        i+=1
        pos_cum[15*i:15+15*i,:]=pos
    elif name=='H2' and previous!='H1':
        pos_new=pos-vec1
        #pos_cum=np.vstack((pos_new,pos))
        pos=pos_new
        previous='H2'
        plt.scatter(pos_new[:,0],pos_new[:,1],s=20,c='b')
        i+=1
        pos_cum[15*i:15+15*i,:]=pos
    elif name=='H3' and previous!='H4':
        pos_new=pos+vec2
        #pos_cum=np.vstack((pos_new,pos))
        pos=pos_new
        previous='H3'
        plt.scatter(pos_new[:,0],pos_new[:,1],s=20,c='k')
        i+=1
        pos_cum[15*i:15+15*i,:]=pos
    elif name=='H4' and previous!='H3':
        pos_new=pos-vec2
        #pos_cum=np.vstack((pos_new,pos))
        pos=pos_new
        previous='H4'
        plt.scatter(pos_new[:,0],pos_new[:,1],s=20,c='g')
        i+=1
        pos_cum[15*i:15+15*i,:]=pos
```

```python
np.savetxt("chain5.xyz",pos_cum)
```
